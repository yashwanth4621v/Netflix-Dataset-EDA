# **Handling** **null** **values**

```python
# For numerical columns â€” fill with median
num_cols = df.select_dtypes(include='number').columns
for col in num_cols:
    if df[col].isnull().sum() > 0:
        median_value = df[col].median()
        df[col] = df[col].fillna(median_value)
```

```python
# For categorical columns â€” fill with mode
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        mode_value = df[col].mode()[0]
        df[col] = df[col].fillna(mode_value)
```

# **updated null values count**

```python
df.isnull().sum()
```

# **handling duplicates**

```python
duplicates = df[df.duplicated()]
print("Duplicate records:")
print(duplicates)
print("Duplicates before removal:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Shape after removing duplicates:", df.shape)
```

```python
df.head()
```

# **extracting month and year from date**

```python

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] =df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

# **Q1: Content Added by year**

```python

year_counts = df['year_added'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
sns.barplot(x=year_counts.index, y=year_counts.values, palette='viridis')
plt.title('Content Added by Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Q2: Top 10 Countires By Content**

```python

top_countries = df['country'].str.split(',').explode().str.strip().value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='coolwarm')
plt.title('Top 10 Countries by Content Count')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

# **Q3: Top 10 genres**

```python

top_genres = df['listed_in'].str.split(',').explode().str.strip().value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='magma')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

# **Q4: Movies vs Tv Show Distribution**

```python

type_counts = df['type'].value_counts()
plt.figure(figsize=(6, 6))
type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#66c2a5', '#fc8d62'])
plt.title('Distribution of Content Types')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

# **Summary**

```python

summary = {
    "Most Active Year": year_counts.idxmax(),
    "Most Titles in a Year": year_counts.max(),
    "Top Country": top_countries.idxmax(),
    "Most Common Genre": top_genres.idxmax(),
    "Content Type Distribution": type_counts.to_dict()
}

print("--- Summary Report ---")
for key, value in summary.items():
    print(f"{key}: {value}")
```
